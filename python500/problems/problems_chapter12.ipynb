```python
# 16
# Check for number is it prime or not.
# for loop.
count = 0
num1 = int(input("Enter your number(positive integer):" ))
for eachNum in range(1, num1+1):    
    if(num1 % eachNum == 0):
        count += 1

if(count == 2):
    print("Prime")
else:
    print("Not prime")

#----------------------------------------
```

```python
# 17 check first 100 number is it prime or not.
# for loop.
num = 1
for i in range(100):
    count = 0
    for j in range(1, num+1):
        if(num % j == 0):
            count += 1
    if(count == 2):
        print("This number is prime:", num)
    num += 1
```

```python
 # 18 first 100th prime.
# for loop.
prime_count = 0
num = 1
while prime_count <= 100:
    prime_indicator = 0
    for i in range(1, num+1):
        if(num % i == 0):
            prime_indicator += 1
    if(prime_indicator == 2):
        print(num)
        prime_count += 1
    num += 1
```
